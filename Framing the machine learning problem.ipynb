# Framing the Machine Learning Problem

# Churn Rate
Churn rate is a metric that shows how many users or customers stop using a product or service during a given time period.

Simple definition

Churn rate = percentage of customers who leave

# Why churn rate matters

- Shows customer satisfaction

- Helps predict revenue loss

- Very important in data science & machine learning (churn prediction models)

# Types of churn

- Customer churn – users stop using the service

- Revenue churn – loss of income

- Voluntary churn – customers choose to leave

- Involuntary churn – payment failure, etc.

# Batch Learning

- Model is trained on the entire dataset at once

- Training happens offline

- Requires more memory

- Model must be retrained when new data arrives

- Suitable for static datasets

- Example: Training a model using last year’s data

# Online Learning

- Model learns one data point or small chunks at a time

- Training happens continuously

- Requires less memory

- Model updates automatically with new data

- Suitable for streaming or dynamic data

- Example: Spam detection system learning from new emails
