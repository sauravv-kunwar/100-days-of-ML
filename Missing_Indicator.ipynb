# Missing Indicator 

- A missing indicator is a binary variable created to flag whether a value is missing in a dataset.

- It is commonly used in statistics, machine learning, and data preprocessing.

- The indicator typically takes values:

- 1 → value is missing

- 0 → value is present

- It is often used alongside imputation (mean, median, zero, etc.), not as a replacement.

# Purpose:

- Allows the model to learn patterns related to missingness itself

- Useful when missing values are informative (not random)

# Example:

- Original variable: Income

- New variable: Income_missing

- If Income is missing → Income_missing = 1

# Advantages:

- Simple and easy to implement

- Helps capture bias introduced by missing data

- Works well with linear models and tree-based models

# Disadvantages:

- Can increase dimensionality

- May introduce noise if missingness is completely random

- Less effective if overused on many variables

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.impute import MissingIndicator,SimpleImputer
```

```python
df = pd.read_csv('train.csv',usecols=['Age','Fare','Survived'])
```

```python
df.head()
```

```python

X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
si = SimpleImputer()
X_train_trf = si.fit_transform(X_train)
X_test_trf = si.transform(X_test)
```

```python
X_train_trf
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(X_train_trf,y_train)

y_pred = clf.predict(X_test_trf)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python

mi = MissingIndicator()

mi.fit(X_train)
```

```python

mi.features_
```

```python
X_train_missing = mi.transform(X_train)
```

```python
X_train_missing
```

```python
X_test_missing = mi.transform(X_test)
```

```python
X_test_missing
```

```python
X_train['Age_NA'] = X_train_missing
```

```python
X_test
```

```python
X_test['Age_NA'] = X_test_missing
```

```python
X_train
```

```python
si = SimpleImputer()

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(X_train_trf2,y_train)

y_pred = clf.predict(X_test_trf2)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
si = SimpleImputer(add_indicator=True)
```

```python
X_train = si.fit_transform(X_train)
```

```python
X_test = si.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(X_train_trf2,y_train)

y_pred = clf.predict(X_test_trf2)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
