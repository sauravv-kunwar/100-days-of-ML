# Principal Component Analysis (PCA)

## âœ… What is PCA?
- **Principal Component Analysis (PCA)** is a **dimensionality reduction technique**.
- It reduces the number of features while keeping maximum useful information.
- PCA transforms original features into **Principal Components (PCs)**.

---

# ðŸ“Œ Importance of Variance and Covariance in PCA

## ðŸ”¸ Variance (Importance)
- **Variance** measures how much a feature spreads from its mean.
- In PCA, **higher variance = more information**.
- PCA selects directions (PCs) that capture **maximum variance**.

### âœ… Variance Formula
For a feature \(X\) with \(n\) observations:

$$
Var(X)=\frac{1}{n}\sum_{i=1}^{n}(x_i-\bar{x})^2
$$

Where:
- \(x_i\) = data value  
- \(\bar{x}\) = mean of the feature  
- \(n\) = number of observations  

---

## ðŸ”¸ Covariance (Importance)
- **Covariance** shows how two variables change together.
- If two features have high covariance, they are correlated and contain similar information.
- PCA uses covariance to combine correlated features into fewer components.

### âœ… Covariance Formula
For two features \(X\) and \(Y\):

$$
Cov(X,Y)=\frac{1}{n}\sum_{i=1}^{n}(x_i-\bar{x})(y_i-\bar{y})
$$

Interpretation:
- \(Cov(X,Y)>0\): both increase/decrease together  
- \(Cov(X,Y)<0\): one increases while other decreases  
- \(Cov(X,Y)=0\): no linear relationship  

---

## ðŸ”¹ Covariance Matrix
For 2 features \(X\) and \(Y\):

$$
C=
\begin{bmatrix}
Var(X) & Cov(X,Y) \\
Cov(Y,X) & Var(Y)
\end{bmatrix}
$$

For multiple features, covariance matrix is an \(m \times m\) matrix.

---

# ðŸ”¹ PCA Mathematical Steps

## 1) Standardize the data
PCA is scale-sensitive, so standardization is important.

---

## 2) Compute Covariance Matrix
If \(X\) is the standardized data matrix:

$$
C=\frac{1}{n}X^T X
$$

---

## 3) Find Eigenvalues and Eigenvectors
Eigen equation:

$$
Cv=\lambda v
$$

Where:
- \(C\) = covariance matrix  
- \(v\) = eigenvector (direction of PC)  
- \(\lambda\) = eigenvalue (importance of PC)

---

## 4) Select top k components
Choose eigenvectors with highest eigenvalues.

---

## 5) Transform data into new space
If \(W\) is the matrix of selected eigenvectors:

$$
Z=XW
$$

Where:
- \(Z\) = new transformed dataset (principal components)

---

# ðŸ”¹ Explained Variance Ratio
The variance captured by each component:

$$
Explained\ Variance\ Ratio=\frac{\lambda_i}{\sum_{j=1}^{m}\lambda_j}
$$

---

## âœ… Advantages of PCA
- Reduces dimensionality
- Removes multicollinearity
- Speeds up model training
- Helps in visualization

---

## âŒ Disadvantages of PCA
- PCs are not easily interpretable
- Some information loss possible
- Best for linear relationships
